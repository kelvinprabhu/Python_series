Stop Iteration Error

```python
from typing import Generator
def count_up_to(max: int) -> Generator[int, None, None]:
    count: int = 0
    while count <= max:
        yield count
        count += 1
my_generator: Generator[int, None, None] = count_up_to(5)
print(next(my_generator))  
```
