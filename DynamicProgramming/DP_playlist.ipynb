Fib in DP

```python
def fib(n,dp = None):
    if dp is None:
        dp = [-1] * (n + 1)
    if n ==0:
        return 0
    elif n==1:
        return 1
    elif dp[n] != -1:
        return dp[n]
    
    else:
        dp[n] = fib(n-1,dp) + fib(n-2,dp)
        return dp[n]
# top down memoization approach -- O(n) time and O(n) space --- dp array of size n+1
```
