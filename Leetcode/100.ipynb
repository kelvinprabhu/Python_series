```python
def kthDigit(a, b, k):
    mod = 10 ** k
    res = 1
    base = a

    while b > 0:
        if b & 1:
            res = (res * base) % mod
        base = (base * base) % mod
        b >>= 1

    for i in range(1, k):
        res //= 10
    return res
```

roman to int

```python
roman_numeral = "MCMXCIV"
def romanToInt(s: str) -> int:
    translations = {"I": 1, "V": 5, "X": 10, "L": 50, "C": 100, "D": 500, "M": 1000}
    total = 0
    prev_value = 0
    s = s.replace("IV","IIII").replace("IX","VIIII").replace("XL","XXXX").replace("XC","LXXXX").replace("CD","CCCC").replace("CM","DCCCC")
    for char in s:
        value = translations[char]
        total += value
    return total
romanToInt(roman_numeral)
```

```python
# 3 sum problem
nums = [-1,0,1,2,-1,-4]
nums.sort()
length = len(nums)
result = []
# print(nums)
for i in range(length):

    for j in range(i+1,length):
        for k in range(j+1,length):
            if nums[i] + nums[j] + nums[k] == 0:
                triplet = sorted([nums[i],nums[j],nums[k]])
                if triplet not in result:
                    result.append(triplet)
print(result)            
                   
            

```

```python
num = 21
binary = bin(num)
# int is shadowed, use __builtins__.int instead
num_int = __builtins__.int(binary, 2)
print(num_int)
print(binary[2:])
arr = [2,3,5]
# Strings are immutable, need to convert to list to modify
binary_list = list(binary[2:])
for n in arr:
    if n <= len(binary_list):
        binary_list[n-1] = '1'
binary_str = ''.join(binary_list)
print(__builtins__.int(binary_str, 2))
```
