# Largest palindrome substring

```python
palindrome_string = "hoelleoooehoooo"
len_palindrome = len(palindrome_string)
low = 1
palindrome_dict = {}
for i in range(len_palindrome):
    for j in range(len_palindrome-1,i,-1):
        # print(i,j)
        sub = palindrome_string[i:j+1]
        if sub == sub[::-1]:
            if palindrome_string[i] == palindrome_string[j]:
                if j-i+1 > low:
                    low = j-i+1
                    palindrome_dict[low] = palindrome_string[i:j+1]
# print(palindrome_string[low-1])
print(palindrome_dict[low])

        
        
        
```

```python
def kthDigit(a, b, k):
    mod = 10 ** k
    res = 1
    base = a

    while b > 0:
        if b & 1:
            res = (res * base) % mod
        base = (base * base) % mod
        b >>= 1

    for i in range(1, k):
        res //= 10
    return res
```

roman to int

```python
roman_numeral = "MCMXCIV"
def romanToInt(s: str) -> int:
    translations = {"I": 1, "V": 5, "X": 10, "L": 50, "C": 100, "D": 500, "M": 1000}
    total = 0
    prev_value = 0
    s = s.replace("IV","IIII").replace("IX","VIIII").replace("XL","XXXX").replace("XC","LXXXX").replace("CD","CCCC").replace("CM","DCCCC")
    for char in s:
        value = translations[char]
        total += value
    return total
romanToInt(roman_numeral)
```
